```python
import numpy as np
import argparse
import cv2
import matplotlib.pyplot as plt
```

```python
def adjust_gamma(image, gamma=1.0):
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    table = np.array([((i / 255.0) ** gamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
    # apply gamma correction using the lookup table
    return cv2.LUT(image, table)
```

```python
img = cv2.imread("Weeki_Wachee_spring_10079u.jpg")
```

```python
img_enhanced = adjust_gamma(img , 2)
```

```python
plt.imshow(img_enhanced[...,::-1])
```

```python
img_enhanced = adjust_gamma(img , 0.5)
plt.imshow(img_enhanced[...,::-1])
```

```python
img_enhanced = adjust_gamma(img , 1/3)
plt.imshow(img_enhanced[...,::-1])
```

```python
img_enhanced = adjust_gamma(img , 1/4)
plt.imshow(img_enhanced[...,::-1])
```

```python
alpha = float(input('Enter the alpha value:[0.1~3]'))
beta = int(input('Enter beta :[0~100]'))
```

```python
new_image = cv2.convertScaleAbs(img , alpha = alpha , beta = beta)
```

```python
plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(img[...,::-1]);plt.title("original")
plt.subplot(122);plt.imshow(new_image[...,::-1]);plt.title("enhanced")
```
